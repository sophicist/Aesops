```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from plotnine import *
```

```python
df = pd.read_csv('global_development.csv')
#os.listdir()

# rename some columns
df = df.rename(columns = {'Data.Health.Life Expectancy at Birth, Female':"LifeExp_Female",
                          "Data.Health.Life Expectancy at Birth, Male":"LifeExp_Male"})
df.head()
```

```python
# African Countries
east_africa = ['Uganda','Tanzania','Kenya','Rwanda']
df_east = df[df.Country.isin(east_africa)]


# ensure that the year is datetime
#df_east.Year  = pd.DatetimeIndex(df_east.Year)
df_east.head()
```

```python
data = pd.melt(df_east,id_vars = ["Country","Year"])
needed = ["LifeExp_Female","LifeExp_Male"]
needed = data[data.variable.isin(needed)]
(ggplot(needed,aes(x = 'Year',y = 'value',color = "Country"))+geom_line()+facet_grid('variable~')
)
```

## The graph shows for example that rwanda had very low life expectancy especially between 1990 - 2000 during the genocide but that has improved and overtaken al the other countries
