```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from plotnine import *
```

```python
df = pd.read_csv('global_development.csv')
#os.listdir()

# rename some columns
df = df.rename(columns = {'Data.Health.Life Expectancy at Birth, Female':"LifeExp_Female",
                          "Data.Health.Life Expectancy at Birth, Male":"LifeExp_Male"})
df.head()
```

```python
# African Countries
east_africa = ['Uganda','Tanzania','Kenya','Rwanda']
df_east = df[df.Country.isin(east_africa)]


# ensure that the year is datetime
#df_east.Year  = pd.DatetimeIndex(df_east.Year)
df_east.head()
```

```python
data = pd.melt(df_east,id_vars = ["Country","Year"])
needed = ["LifeExp_Female","LifeExp_Male"]
needed = data[data.variable.isin(needed)]
(ggplot(needed,aes(x = 'Year',y = 'value',color = "Country"))+geom_line()+facet_grid('variable~')
)
```

## The graph shows for example that rwanda had very low life expectancy especially between 1990 - 2000 during the genocide but that has improved and overtaken al the other countries

# Difference in populations

```python
df_east.head()
```

```python
years = list(range(2010,2014))
df_test = df_east[df_east.Year.isin(years)]
df_test =df_test.fillna(0)
print((ggplot(df_test,aes(x = 'Year',y = 'Data.Health.Total Population',color = "Country"))
 +geom_line()+labs(title ="East African Countries ppn")))

pd.pivot_table(df_test[['Country',"Year","Data.Health.Total Population"]],
               index = ["Year"],columns =['Country'],values = ["Data.Health.Total Population"])
```

```python
cors = df_east[['Data.Health.Death Rate',"Data.Health.Population Growth","LifeExp_Female"]].corr()
#cors
import seaborn as sns
%matplotlib inline

plt.figure()
# plot the heatmap
sns.heatmap(cors, 
        xticklabels=cors.columns,
        yticklabels=cors.columns)
plt.title("Correlations",color ="red",size = 16)
plt.show()
cors
```
