```python
import os
import pandas as pd
import nltk
```

```python
text = pd.DataFrame(columns = ["text","Sentiment"])
text["text"] = [
                "He is the nicing  guy",
                "john went to the market",
                "you are an idiot",
                "He is a male stripper",
                "she did not like him very much",
                "I have gotten a job",
                "there was war after elections",
                "Marving sleeps until late in the morning",
                "you have ugly legs",
                "The president is now fat"
                ]
text["Sentiment"] = ["Positive","Neutral","Negative","Negative","Negative","Positive","Negative","Neutral","Negative","Negative"]
text.head()
```

```python
# word tokenization
from string import punctuation
# nltk.download('punkt')
nltk.download('stopwords')
stop = nltk.corpus.stopwords.words("english")+list(punctuation)
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer 





def cleaner(sentence):
  # tokenize
  clean = nltk.word_tokenize(sentence)
  # remove stopwords
  clean = [i for i in clean if not i in stop]
  # lematize
  lemmatizer = WordNetLemmatizer()
  clean = [lemmatizer.lemmatize(i) for i in clean]

  return " ".join(clean)

cleaner(text["text"][0])
# remove stopwords
# stemming
# lematization
# n-grams analysis
# visualization

text["text"] = list(map(cleaner,text["text"]))
text.head()
```

```python
# visualize the data
## use a frequency chart
## word cloud
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

x = text["text"]
y = text["Sentiment"]

# vectorize the data (learn more about vectorization and bag of words)

vector = CountVectorizer()
X = vector.fit_transform(x)
print(vector.get_feature_names_out())
X.toarray()

dy = pd.DataFrame(X.toarray(),columns = vector.get_feature_names_out())
dy.head()
```

```python
x_train, x_test,y_train,y_test = train_test_split(X,y)
model =  MultinomialNB()
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
print(accuracy_score(pred,y_test))
```
